```python
# web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4

## load, hunk and index the content of the html page
loader=WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
                     bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                         class_=("post-title","post-content","post-header") 
                     )),)

text_documets=loader.load()
```

```python
text_documets
```

```python
## PDF reader-----> Load
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader('Outcome.pdf')
docs=loader.load()
```

```python
docs
```

```python
## split -------> Transform
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
documnets=text_splitter.split_documents(docs)
documnets[:5]
```

```python
documnets[:2]
```

```python
## Vector Embedding And Vector Strore ------->Embed
# from langchain_community.embeddings import OpenAIEmbeddings
import os
os.environ['HTTP_PROXY'] = "http://127.0.0.1:7890"
os.environ['HTTPS_PROXY'] = "http://127.0.0.1:7890"

import requests
try:
    response = requests.get("https://huggingface.co", timeout=5)
    print("网络连接正常")
except:
    print("网络连接异常")

from langchain_community.embeddings import HuggingFaceBgeEmbeddings


model_name = "BAAI/bge-large-zh-v1.5"
model_kwargs = {'device': 'cpu', "trust_remote_code":True}
encode_kwargs = {'normalize_embeddings': True}
hf_emb_model = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)

## 官方的endpoint服务要钱
# from langchain_huggingface.embeddings import HuggingFaceEndpointEmbeddings
# hf_api_embeddings = HuggingFaceEndpointEmbeddings(
#     huggingfacehub_api_token=huggingfacehub_api_token,
#     task="feature-extraction",
#     # model_name="BAAI/bge-large-zh-v1.5",# )
# hf_api_embeddings.embed_query("asdfasdf")

from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(documnets[:40], hf_emb_model)
# db=Chroma.from_documents(documnets[:4], hf_api_embeddings)
```

```python
## Vector DataBase
query="show me the table of contents"
result=db.similarity_search(query)
print(result)
```
